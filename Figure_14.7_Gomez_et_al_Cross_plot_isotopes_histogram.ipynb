```python
d13C=datos.iloc[:, 0]
d18O=datos.iloc[:, 1]
```

```python
plt.scatter(d18O, d13C)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="white", color_codes=True)

g = sns.jointplot(x="d18O", y="d13C", data=datos)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(d18O,d13C)

sns.set(style="darkgrid")
g = sns.jointplot("d18O", "d13C", data=datos, kind="reg")

print('slope=', slope, 'intercept', intercept, 'r_value=', r_value, 'p_value=', p_value, 'std_err=', std_err)
print("y={0:.1f}x+{1:.1f}".format(slope,intercept))

g.savefig("d13C_d18O_histograms.pdf")
```
